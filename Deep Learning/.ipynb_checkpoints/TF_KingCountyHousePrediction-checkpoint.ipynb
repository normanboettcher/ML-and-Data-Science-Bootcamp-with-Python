## Feature Engineering Process

```python
df.head()
```

```python
df= df.drop('id', axis=1)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['date']
```

```python
df['year'] = df['date'].apply(lambda date: date.year)
df['month'] = df['date'].apply(lambda date: date.month)
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='month', y = 'price', data= df)
```

<p> Boxplot does not show useful information after first step </p>

```python
df.groupby('month').mean()['price'].plot()
```

```python
df.groupby('year').mean()['price'].plot()
```

```python
df = df.drop('date', axis=1) #dont need date anymore
```

```python
df['zipcode'].value_counts()
```

```python
df = df.drop('zipcode', axis=1)
```

```python
df['yr_renovated'].value_counts()
```

<p> a lot of houses are not renovated. So we can put that feature in a binary 0 if not renovated and 1 if renovated </p>

```python
df['yr_renovated'] = df['yr_renovated'].apply(lambda value: 1 if value > 0  else 0)
```

```python
df['yr_renovated'].value_counts()
```

```python
df['sqft_basement'].value_counts()
```

<p>we handle sqft_basement as continuous</p>

```python
X = df.drop('price', axis=1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test) #dont fit to test set
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape # 19 features as input.
```

```python
model = Sequential()

model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test),
         batch_size=128)
```
